# Logistic Regression vs. SVM: A Comparison
In this notebook, we will implement and compare logistic regression and SVM (Support Vector Machine) using a simple dataset to understand their performance differences.

We'll start by importing necessary libraries and loading the dataset.


```python
# Step 1: Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris  # Using a simple dataset as an example
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Step 2: Load a binary classification dataset
# Using Iris dataset for simplicity; we'll select only two classes for binary classification.
iris = load_iris()
X = iris.data
y = iris.target

# Filter to only two classes for binary classification (e.g., class 0 and 1)
X = X[y != 2]
y = y[y != 2]

# Step 3: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Logistic Regression Implementation
We will first train a logistic regression model and evaluate its performance on the test data.


```python
# Step 5: Train and evaluate Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Predict using the logistic regression model
y_pred_log_reg = log_reg.predict(X_test)

# Evaluate the model's performance
print("Logistic Regression Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_log_reg))
print("Classification Report:\n", classification_report(y_test, y_pred_log_reg))
```

## SVM Implementation
Now, we will train an SVM model with a linear kernel and compare its performance with logistic regression.


```python
# Step 6: Train and evaluate SVM
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

# Predict using the SVM model
y_pred_svm = svm_model.predict(X_test)

# Evaluate the model's performance
print("SVM Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print("Classification Report:\n", classification_report(y_test, y_pred_svm))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # Take the first two features for 2D visualization
y = iris.target

# Filter to two classes for binary classification
X = X[y != 2]
y = y[y != 2]

# Train Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X, y)

# Train SVM with RBF Kernel
svm_rbf = SVC(kernel='rbf', C=1, gamma=0.5)
svm_rbf.fit(X, y)

# Function to plot decision boundary
def plot_decision_boundary(model, X, y, title):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, alpha=0.6, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
    plt.title(title)
    plt.xlabel(iris.feature_names[0])
    plt.ylabel(iris.feature_names[1])
    plt.show()

# Plot decision boundaries for both models
plot_decision_boundary(log_reg, X, y, "Logistic Regression Decision Boundary")
plot_decision_boundary(svm_rbf, X, y, "SVM with RBF Kernel Decision Boundary")
```

```python
# Step 1: Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris, load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Step 2: Load and understand the datasets
iris = load_iris()
breast_cancer = load_breast_cancer()

# Convert datasets to DataFrames
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

breast_cancer_df = pd.DataFrame(data=breast_cancer.data, columns=breast_cancer.feature_names)
breast_cancer_df['target'] = breast_cancer.target

# Display initial data information
print("Iris Dataset:")
print(iris_df.head(), "\n")

print("Breast Cancer Dataset:")
print(breast_cancer_df.head(), "\n")

# Step 3: Data Preparation
# Split data into training and testing sets for both datasets
X_iris = iris_df.drop('target', axis=1)
y_iris = iris_df['target']

X_breast = breast_cancer_df.drop('target', axis=1)
y_breast = breast_cancer_df['target']

X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)
X_train_breast, X_test_breast, y_train_breast, y_test_breast = train_test_split(X_breast, y_breast, test_size=0.3, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_iris = scaler.fit_transform(X_train_iris)
X_test_iris = scaler.transform(X_test_iris)
X_train_breast = scaler.fit_transform(X_train_breast)
X_test_breast = scaler.transform(X_test_breast)

# Step 4: Modeling
# Logistic Regression
log_reg_iris = LogisticRegression(max_iter=200)
log_reg_breast = LogisticRegression(max_iter=200)

log_reg_iris.fit(X_train_iris, y_train_iris)
log_reg_breast.fit(X_train_breast, y_train_breast)

# SVM
svm_iris = SVC(kernel='linear')
svm_breast = SVC(kernel='linear')

svm_iris.fit(X_train_iris, y_train_iris)
svm_breast.fit(X_train_breast, y_train_breast)

# Step 5: Evaluation
# Logistic Regression Evaluation
y_pred_log_iris = log_reg_iris.predict(X_test_iris)
y_pred_log_breast = log_reg_breast.predict(X_test_breast)

print("Logistic Regression - Iris Dataset")
print("Accuracy:", accuracy_score(y_test_iris, y_pred_log_iris))
print("Classification Report:\n", classification_report(y_test_iris, y_pred_log_iris))

print("Logistic Regression - Breast Cancer Dataset")
print("Accuracy:", accuracy_score(y_test_breast, y_pred_log_breast))
print("Classification Report:\n", classification_report(y_test_breast, y_pred_log_breast))

# SVM Evaluation
y_pred_svm_iris = svm_iris.predict(X_test_iris)
y_pred_svm_breast = svm_breast.predict(X_test_breast)

print("SVM - Iris Dataset")
print("Accuracy:", accuracy_score(y_test_iris, y_pred_svm_iris))
print("Classification Report:\n", classification_report(y_test_iris, y_pred_svm_iris))

print("SVM - Breast Cancer Dataset")
print("Accuracy:", accuracy_score(y_test_breast, y_pred_svm_breast))
print("Classification Report:\n", classification_report(y_test_breast, y_pred_svm_breast))

# Step 6: Deployment (conceptual, not shown in code)
# This is where the model would be integrated into an application or further tested in a production environment.
```
